## Scatter Plot

```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
df1=pd.read_csv("googleplaystore.csv",nrows=1000)
```

```python
df1
```

```python
df1.shape
```

```python
df1.head()
```

```python
x=df1["Rating"]
y=df1["Reviews"]
plt.scatter(x,y)
plt.show()
```

1. xlabel,ylabel,title

```python
x=df1["Rating"]
y=df1["Reviews"]
plt.scatter(x,y)
plt.title("Google Play Store Apps Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Review")
plt.show()
```

Parameters

1. color

```python
plt.scatter(x,y,color='r')
plt.title("Google Play Store Apps Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Review")
plt.show()
```

2. Increase the size of scatter

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y,color='r')
plt.title("Google Play Store Apps Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Review")
plt.show()
```

3. Marker

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y,color='r',marker="*")
plt.title("Google Play Store Apps Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Review")
plt.show()
```

4. Increase the size of marker

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y,color='r',marker="*",s=150)
plt.title("Google Play Store Apps Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Review")
plt.show()
```

5. alpha

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y,color='r',marker="*",s=150,alpha=0.5)
plt.title("Google Play Store Apps Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Review")
plt.show()
```

6. linewidth

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y,color='r',marker="*",s=100,alpha=0.5,linewidth=10)
plt.title("Google Play Store Apps Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Review")
plt.show()
```

7. change the color of linewidth

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y,color='r',marker="*",s=100,alpha=0.5,linewidth=10,edgecolor='g')
plt.title("Google Play Store Apps Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Review")
plt.show()
```

8. verts which is work as similar to marker

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y,color='r',s=100,alpha=0.5,linewidth=10,edgecolor='g',verts="<")
plt.title("Google Play Store Apps Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Review")
plt.show()
```

9. Plot two scatter in one

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y,color='r',marker="*",s=100,alpha=0.5,linewidth=10,edgecolor='g')
plt.scatter(x,df1["Installs"],color='b',marker="o",s=100,alpha=0.5,linewidth=10,edgecolor='m')
plt.title("Google Play Store Apps Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Reviews and Installs")
plt.show()
```
