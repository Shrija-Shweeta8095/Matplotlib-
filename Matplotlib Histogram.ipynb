### Histogram

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
import random
```

```python
ml_student_age=np.random.randint(18,45,(100))
py_student_age=np.random.randint(15,40,(100))
```

```python
print(ml_student_age)
print(py_student_age)
```

```python
plt.hist(ml_student_age)
plt.show()
```

```python
plt.hist(ml_student_age)
plt.title("ML students age histogram")
plt.xlabel("Students age category")
plt.ylabel("No. of students age")
plt.show()
```

#### Parameters

1. bins

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_student_age,bins)
plt.title("ML students age histogram")
plt.xlabel("Students age category")
plt.ylabel("No. of students age")
plt.show()
```

2. rwidth

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_student_age,bins,rwidth=0.8)
plt.title("ML students age histogram")
plt.xlabel("Students age category")
plt.ylabel("No. of students age")
plt.show()
```

3.histtype

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_student_age,bins,rwidth=0.8,histtype="step")
plt.title("ML students age histogram")
plt.xlabel("Students age category")
plt.ylabel("No. of students age")
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_student_age,bins,rwidth=0.8,histtype="bar")
plt.title("ML students age histogram")
plt.xlabel("Students age category")
plt.ylabel("No. of students age")
plt.show()
```

4. align- left, mid,right

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_student_age,bins,rwidth=0.8,histtype="bar",align="left")
plt.title("ML students age histogram")
plt.xlabel("Students age category")
plt.ylabel("No. of students age")
plt.show()
```

5. orientation- horizontal, vertical

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_student_age,bins,rwidth=0.8,histtype="bar",align="left",orientation="horizontal")
plt.title("ML students age histogram")
plt.xlabel("Students age category")
plt.ylabel("No. of students age")
plt.show()
```

6. color

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_student_age,bins,rwidth=0.8,histtype="bar",align="left",orientation="horizontal",color='r')
plt.title("ML students age histogram")
plt.xlabel("Students age category")
plt.ylabel("No. of students age")
plt.show()
```

7. label

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_student_age,bins,rwidth=0.8,histtype="bar",align="left",orientation="horizontal",color='r',label="ML students")
plt.title("ML students age histogram")
plt.xlabel("Students age category")
plt.ylabel("No. of students age")
plt.legend()
plt.show()
```

8. Combining two information in one histogram

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_student_age,bins,rwidth=0.8,histtype="bar",align="left",orientation="vertical",color='r',label="ML students")
plt.hist(py_student_age,bins,rwidth=0.8,histtype="bar",align="left",orientation="vertical",color='g',label="Python Students")
plt.title("ML students age histogram")
plt.xlabel("Students age category")
plt.ylabel("No. of students age")
plt.legend()
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist([ml_student_age,py_student_age],bins,rwidth=0.8,histtype="bar",align="left",orientation="vertical",color=['r','g'],label=["ML students","Python Student"])
#plt.hist(py_student_age,bins,rwidth=0.8,histtype="bar",align="left",orientation="vertical",color='g',label="Python Students")
plt.title("ML students age histogram")
plt.xlabel("Students age category")
plt.ylabel("No. of students age")
plt.legend()
plt.show()
```

Increase the size of histogram

```python
bins=[15,20,25,30,35,40,45]
plt.figure(figsize=(16,9))
plt.hist([ml_student_age,py_student_age],bins,rwidth=0.8,histtype="bar",align="left",orientation="vertical",color=['r','g'],label=["ML students","Python Student"])
#plt.hist(py_student_age,bins,rwidth=0.8,histtype="bar",align="left",orientation="vertical",color='g',label="Python Students")
plt.title("ML students age histogram")
plt.xlabel("Students age category")
plt.ylabel("No. of students age")
plt.legend()
plt.show()
```

Change the background color
