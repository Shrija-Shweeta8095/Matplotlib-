## Pie Chart

```python
import matplotlib.pyplot as plt
```

```python
classes=['Python','R','Machine Learning','Artificial Intelligence','Data Science']
class1_students=[45,15,35,25,30]
```

```python
plt.pie([1])
plt.show()
```

```python
plt.pie([class1_students])
plt.show()
```

1. Label

```python
plt.pie([class1_students],labels=classes)
plt.show()
```

2. explode- to show anything in slices

```python
explode=[0,0,0,0,0]
plt.pie([class1_students],labels=classes,explode=explode)
plt.show()
```

```python
explode=[0.03,0,0.1,0,0]
plt.pie([class1_students],labels=classes,explode=explode)
plt.show()
```

3. color

```python
explode=[0.03,0,0.1,0,0]
color=["c","r","g","b","y"]
plt.pie([class1_students],labels=classes,explode=explode,colors=color)
plt.show()
```

4. autopct- to show value in %

```python
explode=[0.03,0,0.1,0,0]
color=["c","r","g","b","y"]
plt.pie([class1_students],labels=classes,explode=explode,colors=color,autopct="%0.2f%%")
plt.show()
```

5. shadow by default False

```python
explode=[0.03,0,0.1,0,0]
color=["c","r","g","b","y"]
plt.pie([class1_students],labels=classes,explode=explode,colors=color,autopct="%0.2f%%",shadow=True)
plt.show()
```

6. radius- to increase the radius of pie chart

```python
explode=[0.03,0,0.1,0,0]
color=["c","r","g","b","y"]
plt.pie([class1_students],labels=classes,explode=explode,colors=color,autopct="%0.2f%%",shadow=True,
       radius=1.4)
plt.show()
```

7. startangle- to plot the 1st value of pie chart according to our own angle

```python
explode=[0.03,0,0.1,0,0]
color=["c","r","g","b","y"]
plt.pie([class1_students],labels=classes,explode=explode,colors=color,autopct="%0.2f%%",shadow=True,
       radius=1.4,startangle=270)
plt.show()
```

8. textprops- text properties
use to change the size, text of the text written on pie chart
Take input in dictionary format

```python
explode=[0.03,0,0.1,0,0]
color=["c","r","g","b","y"]
textprops={'fontsize':15}
plt.pie([class1_students],labels=classes,explode=explode,colors=color,autopct="%0.2f%%",shadow=True,
       radius=1.4,startangle=270,textprops=textprops)
plt.show()
```

9. counterclock- in which direction you want to plot your pie chart

```python
explode=[0.03,0,0.1,0,0]
color=["c","r","g","b","y"]
textprops={'fontsize':15}
plt.pie([class1_students],labels=classes,explode=explode,colors=color,autopct="%0.2f%%",shadow=True,
       radius=1.4,startangle=270,textprops=textprops,counterclock=False)
plt.show()
```

10. wedgeprops- to change the size of pie, change the length of pie, change the color of edge

```python
explode=[0.03,0,0.1,0,0]
color=["c","r","g","b","y"]
textprops={'fontsize':15}
wedgeprops={"linewidth":4,"width":1,"edgecolor":"k"}
plt.pie([class1_students],labels=classes,explode=explode,colors=color,autopct="%0.2f%%",shadow=True,
       radius=1.4,startangle=270,textprops=textprops,wedgeprops=wedgeprops)
plt.show()
```

11. center- where to plot pie chart on x and y axis

```python
explode=[0.03,0,0.1,0,0]
color=["c","r","g","b","y"]
textprops={'fontsize':15}
wedgeprops={"linewidth":4,"width":1,"edgecolor":"k"}
plt.pie([class1_students],labels=classes,explode=explode,colors=color,autopct="%0.2f%%",shadow=True,
       radius=1.4,startangle=270,textprops=textprops,wedgeprops=wedgeprops,center=(2,3))
plt.show()
```

```python
explode=[0.03,0,0.1,0,0]
color=["c","r","g","b","y"]
textprops={'fontsize':15}
wedgeprops={"linewidth":4,"width":1,"edgecolor":"k"}
plt.pie([class1_students],labels=classes,explode=explode,colors=color,autopct="%0.2f%%",shadow=True,
       radius=1.6,startangle=270,textprops=textprops,wedgeprops=wedgeprops,center=(1,3),frame=True,
       rotatelabels=True)
plt.show()
```

```python
explode=[0.03,0,0.1,0,0]
color=["c","r","g","b","y"]
textprops={'fontsize':15}
wedgeprops={"linewidth":4,"width":2,"edgecolor":"k"}
plt.pie([class1_students],labels=classes,explode=explode,colors=color,autopct="%0.2f%%",shadow=True,
       radius=1.6,startangle=270,textprops=textprops,wedgeprops=wedgeprops,center=(1,3),frame=True,
       rotatelabels=True)
plt.show()
```

```python
plt.figure(figsize=(2,2))
explode=[0.03,0,0.1,0,0]
color=["c","r","g","b","y"]
textprops={'fontsize':15}
wedgeprops={"linewidth":4,"width":3,"edgecolor":"k"}
plt.pie([class1_students],labels=classes,explode=explode,colors=color,autopct="%0.2f%%",shadow=True,
       radius=1.6,startangle=270,textprops=textprops,wedgeprops=wedgeprops,center=(1,3),frame=True,
       rotatelabels=True)
plt.show()
```

legend

```python
explode=[0.03,0,0.1,0,0]
color=["c","r","g","b","y"]
textprops={'fontsize':15}
wedgeprops={"linewidth":4,"width":1,"edgecolor":"k"}
plt.pie([class1_students],labels=classes,explode=explode,colors=color,autopct="%0.2f%%",shadow=True,
       radius=1.4,startangle=270,textprops=textprops,wedgeprops=wedgeprops)
plt.legend()
plt.show()
```
