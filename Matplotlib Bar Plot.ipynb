## Bar Chart

```python
from matplotlib import pyplot as plt
import numpy as np
from matplotlib import style
```

```python
## Datasets
classes=["Python","R","ML","AI","DS"]
class1_students=[30,10,25,20,10]
class2_students=[40,5,20,20,10]
class3_students=[35,5,30,25,15]
```

1. bar()

```python
plt.bar(classes,class1_students)
```

2.barh

```python
plt.barh(classes,class1_students)
```

Parameters

1. width

```python
plt.bar(classes,class1_students,width=0.2)
```

2. align by default center

```python
plt.bar(classes,class1_students,width=0.2,align="edge")
```

3. color

```python
plt.bar(classes,class1_students,width=0.2,align="edge",color="r")
```

4. change the color of edge

```python
plt.bar(classes,class1_students,width=0.2,align="edge",color="r",edgecolor='b')
```

5. change the width of line

```python
plt.bar(classes,class1_students,width=0.2,align="edge",color="r",edgecolor='b',linewidth=5)
```

6. change the alpha of red
alpha change the opacity i.e transparency

```python
plt.bar(classes,class1_students,width=0.2,align="edge",color="r",edgecolor='b',linewidth=5,alpha=0.4)
```

7. change the linestyle

```python
plt.bar(classes,class1_students,width=0.2,align="edge",color="r",edgecolor='b',linewidth=3,linestyle="--")
```

8. label

```python
plt.bar(classes,class1_students,width=0.2,align="edge",color="r",edgecolor='b',linewidth=3,linestyle="--",label="Class 1 Students")
plt.legend()
```

9. to hide the graph use visible =False

```python
plt.bar(classes,class1_students,width=0.2,align="edge",color="r",edgecolor='b',linewidth=3,linestyle="--",label="Class 1 Students",visible=False)
plt.legend()
```

10. CHANGE THE WIDTH OF BAR

```python
plt.bar(classes,class1_students,width=0.6,align="edge",color="r",edgecolor='b',linewidth=3,linestyle="--",label="Class 1 Students")
plt.legend()
```

Styles

1. Change the background color of bar plot

```python
style.use('ggplot')
plt.bar(classes,class1_students,width=0.2,align="edge",color="r",edgecolor='b',linewidth=3,linestyle="--",label="Class 1 Students")
plt.legend()
```

2. Increase the size of bar chart

```python
style.use('ggplot')
plt.figure(figsize=(16,9))
plt.bar(classes,class1_students,width=0.2,align="edge",color="r",edgecolor='b',linewidth=3,linestyle="--",label="Class 1 Students")
plt.legend()
```

3. plot bar horizontal

```python
style.use('ggplot')
plt.barh(classes,class1_students,align="edge",color="r",edgecolor='b',linewidth=3,linestyle="--",label="Class 1 Students")
plt.legend()
```

4. title,xlabel,ylabel

```python
style.use('ggplot')
plt.bar(classes,class1_students,align="edge",color="r",edgecolor='b',linewidth=3,linestyle="--",label="Class 1 Students")
plt.title("Bar chart of Classes")
plt.xlabel("Classes")
plt.ylabel("No. of students")
plt.legend()
plt.show()
```

5. Combine all

```python
style.use('ggplot')
class_index=np.arange(len(classes))
width=0.2
plt.bar(class_index,class1_students,align="edge", width=0.2, color="r",label="Class 1 Students")
plt.bar(class_index+width,class2_students,align="edge", width=0.2, color="b",label="Class 2 Students")
plt.xticks(class_index + width,classes,rotation=20)
plt.title("Bar chart of Classes")
plt.xlabel("Classes")
plt.ylabel("No. of students")
plt.legend()
plt.show()
```

```python
style.use('ggplot')
class_index=np.arange(len(classes))
width=0.2
plt.bar(class_index,class1_students,align="edge", width=0.2, color="r",label="Class 1 Students")
plt.bar(class_index+width,class2_students,align="edge", width=0.2, color="b",label="Class 2 Students")
plt.bar(class_index+width+width,class3_students,align="edge", width=0.2, color="m",label="Class 3 Students")
plt.xticks(class_index + width,classes,rotation=20)
plt.title("Bar chart of Classes")
plt.xlabel("Classes")
plt.ylabel("No. of students")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(16,9))
class_index=np.arange(len(classes))
width=0.05
plt.barh(class_index,class1_students,align="edge",  color="r",label="Class 1 Students")
plt.barh(class_index+width,class2_students,align="edge", color="b",label="Class 2 Students")
plt.barh(class_index+width+width,class3_students,align="edge",  color="m",label="Class 3 Students")
plt.yticks(class_index + width,classes,rotation=20)
plt.title("Bar chart of Classes")
plt.ylabel("Classes")
plt.xlabel("No. of students")
plt.legend()
plt.show()
```
